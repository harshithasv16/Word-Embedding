```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
###sentences
sent = ["I am a good girl",
       "I love my mother the most",
       "my brother is super naughty",
       "my dad is super hero",
       "the cup of tea"]
```

```python
sent
```

```python
voc_size = 10000
```

# One hot Representations

```python
#This gives the index from the dictionary
onehot_repr = [one_hot(words,voc_size) for words in sent]
print(onehot_repr)
```

# word embedding representation

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length = 7
embedded_docs = pad_sequences(onehot_repr,padding="pre",maxlen = sent_length)
embedded_docs
```

```python
##dimensions is features
dim= 10
```

```python
##embedding converts into featurized representation using dimensions
model = Sequential()
model.add(Embedding(voc_size, dim, input_length = sent_length))
model.compile("adam",'mse')
#adam optimizer
#mse performance metrics
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
